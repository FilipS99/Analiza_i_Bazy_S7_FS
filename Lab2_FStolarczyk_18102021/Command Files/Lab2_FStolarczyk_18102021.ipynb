```python
import pandas as pd
import numpy as np
```

# Processing the data
Reading csv file, extracting desired columns, renaming them and dropping incomplete answers.

```python
df = pd.read_csv("../Original Data/earthquake_data.csv")
df.head()
```

```python
df_extracted = df[['Age','What is your gender?', 'Do you think the "Big One" will occur in your lifetime?']]
df_extracted.columns = ['age', 'sex', 'answer']
df_extracted = df_extracted.dropna(axis=0)
df_extracted.head()
```

# Constructing the Data Appendix
Splitting data into desired DataFrames

```python
positive_answers = df_extracted.loc[df_extracted['answer']=='Yes']
positive_answers.columns = ['age', 'sex', 'positive answers']
positive_answers = positive_answers.groupby(['age','sex']).count()

negative_answers = df_extracted.loc[df_extracted['answer']=='No']
negative_answers.columns = ['age', 'sex', 'negative answers']
negative_answers = negative_answers.groupby(['age','sex']).count()

total_answers = positive_answers['positive answers']/(positive_answers['positive answers']+negative_answers['negative answers'])
```

# Generating the results
Creating DataFrame for analysis and saving as .xlsx file

```python
result = pd.concat([positive_answers, negative_answers, total_answers.rename('yes/no ratio')], axis=1)
result
```

```python
result.to_excel("../Analysis Data/result.xlsx")
```
